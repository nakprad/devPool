```python
#import modules
import numpy as np
from pyspark.sql.types import *
from pyspark.sql.functions import *

# start
# sc = SparkContext(master="local[4]")
from pyspark.sql import SparkSession
from pyspark import SparkContext
from pyspark import StorageLevel
appName = "hat"
spark = SparkSession.builder.master("local[*]") \
        .appName(appName) \
        .getOrCreate()
spark
```

```python
Schema = StructType([
  StructField("name", StringType()),
  StructField("surname", StringType()),
  StructField("f1", IntegerType()),
  StructField("l1", IntegerType()),
  StructField("sex", StringType()),
  StructField("s1", IntegerType()),
  StructField("smoker", StringType()),
  StructField("s2", IntegerType()),
  StructField("age", IntegerType()),
  StructField("class", StringType()),
  StructField("s3", IntegerType()),
])

csv = spark.read.csv('C:/Users/nakprad/Desktop/konn.csv',schema=Schema, header=True)
```

```python
csv.groupby("class").count().show()
```

```python
csv2 = csv.drop('sex','smoker','class')
```

```python
csv2.collect()
```

```python
dividedData = csv2.randomSplit([0.8, 0.2])
trainingData = dividedData[0] #index 0 = data training
testingData = dividedData[1] #index 1 = data testing
train_rows = trainingData.count()
test_rows = testingData.count()
print ("Training data rows:", train_rows, "; Testing data rows:", test_rows)
```

```python
testingData.groupby("s3").count().show()
```

```python
trainlist = []
for i in trainingData.collect():
    trainlist.append(tuple(i))
```

```python
testlist = []
for j in testingData.collect():
    testlist.append(tuple(j))
```

```python
import math
import time
import numpy as np
import pandas as pd

start_time = time.time()
k = 5
mydict = {}
mydict3 = []
mydict4 = []
mydict5 = []
mydict6 = []
mydict7 = []

for m in range(0,len(testlist)): #each test row === len(testlist) ===
    for j in range(0,len(trainlist)): #each train row ==== len(trainlist) ====
        distance = 0
        distance = math.sqrt(np.square((np.array(testlist[m][2:-1]) - np.array(trainlist[j][2:-1]))).sum())
        upd = {distance:trainlist[j][5]}
        mydict.update(upd)  
    result = sorted(mydict.items(), key = lambda x:x[0])[:k]
#     print(result)
    mydict.clear()
    total = 0
    
    for i in range(0,k):
        total += result[i][1]
#         print(total)
    if total >= 3 and mydict7.count("Gryffindor") <25:
        predict = "Gryffindor"
    elif total >= 2 and mydict7.count("Ravenclaw") <25:
        predict = "Ravenclaw"
    elif total >= 1 and mydict7.count("Slytherin") <25:
        predict = "Slytherin"  
    elif mydict7.count("Hufflepuff") <25:
        predict = "Hufflepuff"
        
        
    actual = testlist[m][5]
    name = testlist[m][0]
    lastname = testlist[m][1]
    mydict3.append(m)
    mydict4.append(name)
    mydict5.append(lastname)
    mydict6.append(total)
    mydict7.append(predict)
    print(m,name,lastname,"อยู่บ้าน",total,predict)
print("finish in ",time.time() - start_time)
```

```python
schema = StructType([StructField("m",IntegerType(),True),StructField("name",StringType(),True),StructField("lastname",StringType(),True),StructField("total",IntegerType(),True),StructField("predict",StringType(),True)])
data1 = list(zip(mydict3,mydict4,mydict5,mydict6,mydict7))
df = spark.createDataFrame(data1,schema)
# # .write.mode('overwrite').format('csv').save('output')
df.toPandas().to_csv('output_konn_k5.csv')
```
